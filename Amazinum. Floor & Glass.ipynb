#### Amazinum. Floor&Glass

drop[1][j] = j, j = 1...N # один стакан, перевірка кожного поверху від 1 до j

drop[i][0] = 0, i = 1...K # немає поверхів, для отримання оптимального поверху кидків не потрібно

drop[i][1] = 1, i = 1...K # один поверх, лишається лише раз перевірити поверх

Наша умова: 2 стакана і 100 поверхів.
Для кидка, я можу вибрати поверх від 1 до 100. Наприклад, я вибрав 25.
Тоді результат буде таким:

1. Стакан б'ється, тоді я маю ще 1 стакан, і повинен перевірити поверхи від 1 до 24.
2. Стакан не б'ється, я все ще маю 2 стакани, і тепер можу вибрати поверхи від 26 до 100.

Проте 25 поверх не підходить, оскільки якщо виберем поверх 25, тоді:
Якщо б'ється, перевіряєм від 1 до 24(24 рази). Якщо не б'ється, вибираєм наступний поверх.

Відповідно наступний поверх буде 49(25+24). І тоді якщо б'ється, то ми втрачаєм всі стакани. Якщо не б'ється, перевіряєм від 26 до 48 поверху. Якщо стакан і далі не б'ється, тоді вибираєм наступні поверхи таким же чином: 72(49+23), 94(72+22). Таким чином кожна перевірка зменшує поверх на 1. Якщо далі вибирати, то буде поверх 115(94+21), якого в нас немає. В цьому варіанті 25 не підходить, оскільки від 94 поверху ще потрібно перевірити поверхи вище, до 100.

Тоді виберу поверх 14 (як нам показує програма). Тоді робим перевірку таку ж саму, як із 25м поверхом і перевіряєм чи підходить цей поверх для першого кидка стакану. 14+13=27 - наступний поверх. Якщо б'ється, перевіряєм від 15 до 27го поверху. Далі 27+12=39. Потім 50, 60, 69, 77, 84, 90, 95, 99(95+4). Відповідно таким чином відбувається перевірка кожного поверху.

```python
class Solution(object):
    def GlassDrop(G, F):
        """
        :type E: int
        :type F: int
        :rtype: int
        """
        
        #float('inf') дія, що перевіряє числа вище
        drop=[[float('inf')]*(F+1) for drop in range(G+1)]
        for i in range(1, G+1):
            drop[i][0] = 0
            drop[i][1] = 1
        for j in range(1, F+1):
            drop[1][j] = j
        
        for i in range(2, G+1):
            for j in range(2, F+1):
                for k in range(1, j+1):
                    drop[i][j] = min(drop[i][j], 1 + max(drop[i-1][k-1], drop[i][j-k]))
        return drop[G][F]

Solution.GlassDrop(2,100)
```

#### 3 стакани і 100 поверхів.
Наша наступна умова: 3 стакана і 100 поверхів.
Для кидка, я можу вибрати поверх від 1 до 100. Наприклад, я вибрав 25.
Тоді результат буде таким:

1. Стакан б'ється, тоді я маю ще 2 стакани, і повинен перевірити поверхи від 1 до 24 (як коли я маю 2 стакани і j поверхів).
2. Стакан не б'ється, я все ще маю 3 стакани, і тепер можу вибрати поверхи від 26 до 100.
В такому випадку отримання оптимального поверху з вибором 25го, з якого кидаєм стакан, можна описати за формулою:
drop[3][100] = 1 + max(drop[2][24], drop[3][75])

Окрім 25го поверху, я можу вибрати любий поверх від 1 до 100. Кожен кидок стакану буде мати результат такий, як при киданні з 25го. Фінальний результат буде мінімум всіх позитивних виборів поверхів для кидку стакану:

drop[3][100] = min(..., 1 + max(drop[2][23], drop[3][76]), 1 + max(drop[2][24], drop[3][75]), 1 + max(drop[2][25], drop[3][74]), ...) (берем поверхи 24, 25, 26 для прикладу)

Загальну формула за прикладом вище буде така:
drop[i][j] = min(1 + max(drop[i-1][k-1], drop[i][j-k])), k = 1,2,...j

```python
class Solution(object):
    def GlassDrop(G, F):
        """
        :type E: int
        :type F: int
        :rtype: int
        """
        
        #float('inf') дія, що перевіряє числа вище
        drop=[[float('inf')]*(F+1) for drop in range(G+1)]
        for i in range(1, G+1):
            drop[i][0] = 0
            drop[i][1] = 1
        for j in range(1, F+1):
            drop[1][j] = j
        
        for i in range(2, G+1):
            for j in range(2, F+1):
                for k in range(1, j+1):
                    drop[i][j] = min(drop[i][j], 1 + max(drop[i-1][k-1], drop[i][j-k]))
        return drop[G][F]

Solution.GlassDrop(3,100)
```

По формулі ми визначаєм за скільки кидків можна досягти 100 поверху. Якщо поверхів 100 і в нас 3 яйця, по формулі ми визначаємо, що кидків 9 має хватити. Тоді ми мусимо найти таке число, при якому якщо лишиться 2 яйця і певна кількість поверхів, то по формулі визначення з 2ма яйцями, треба досягнути 8 кидків. Таке число буде в нас 37. Тобто, перший кидок з 3ма яйцями може бути провальний, і тоді буде перевірятись 36 поверхів із 2ма яйцями. Це буде число 8 при цьому. Тоді знаходимо, скільки поверхів треба, аби кидків було 7 із 2-ма яйцями: 36-8=28. Це число 28. А поверх, з якого кидатиметься саме перше яйце із трьох це 28+38=66 поверх, з якого кидаєм (38 поверх це якщо стакан не розбився з 2ма яйцями). Якщо яйце б'ється, перевіряються 28 поверхів. Далі 28-7=21. 21 за 6 кидків. Тоді 21+67=88. Далі до 100 числа в нас лишиться 11 поверхів, які перевіряться 2ма яйцями за 5 кидків. Тоді 11+89=100. Таким чином перевіряться всі поверхи.
